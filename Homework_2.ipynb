```
import math
from sympy import *
import numpy as np
import matplotlib.pyplot as plt 
```

```
 # Khoi tao x ban dau, l_rate
x = float(input("Khoi tao x tuy y:"))
l_rate = float(input("Nhap learning rate:"))

# Dao ham y = x**2
dy = 2*x

#. Tim Min
while abs(dy) > 10**(-5):
  x = x - l_rate * dy
  dy = 2*x
print("Gia tri mmin y =:",x**2, "tai x = ",x)
```

```
# Ve do thi ham LOSS =  0.5*(ymin_predict-y_min_real)**2
# Khoi tao lai gia tri x0 vs l_rate
x = float(input(" Nhap x ban dau:"))
l_rate = float(input("Nhap l_rate:"))
dy = 2*x
LOSS = []
i = 0
while abs(dy) > 10**(-5):
  x = x - l_rate * dy
  loss_value = 0.5*((x**2)**2)
  LOSS.append(loss_value)
  dy = 2*x
  i += 1
  
plt.plot(LOSS)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show

print("Epoch = ",i)
print("LOSS values:",LOSS)
```
